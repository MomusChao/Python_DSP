```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
```

```python
# time array
t = np.linspace( 0, 1, 1000 )
```

```python
# create Square
x = signal.square( 2 * np.pi * 5 * t )
```

```python
# plotting
plt.plot(t,x)
plt.xlabel('t(second)')
plt.ylabel('Amplitude')
plt.axis([ 0, 1, -1.2, 1.2 ])
plt.show( )
```
