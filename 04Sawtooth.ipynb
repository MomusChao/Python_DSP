```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
```

```python
# time array
t = np.linspace(0,1,1000,endpoint=False)
```

```python
# create sawtooth
x = signal.sawtooth(2*np.pi*5*t)
```

```python
# plotting
plt.plot(t,x)
plt.xlabel('t(second)')
plt.ylabel('Amplitude')
plt.axis([ 0, 1, -1.2, 1.2 ])
plt.show()
```
