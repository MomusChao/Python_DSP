```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# time array: [0 ,0.001 ,0.002 ,... ,0.999]
# endpoint=False: exclude 1
# 1000: interval 1/1000
t = np.linspace(0,1,1000,endpoint=False)
```

```python
# create sinusoids
x = np.cos(2*np.pi*5*t)
```

```python
plt.plot(t,x)
plt.xlabel('t(second)')
plt.ylabel('Amplitude')
plt.show()
```

# Different Amplitude

```python
x2 = 2*np.cos(2*np.pi*5*t)
```

```python
plt.plot(t,x2)
plt.xlabel('t(second)')
plt.ylabel('Amplitude')
plt.show()
```

# Time Delay

```python
x3 = np.cos(2*np.pi*5*t-2)
```

```python
plt.plot(t,x3)
plt.xlabel('t(second)')
plt.ylabel('Amplitude')
plt.show()
```
