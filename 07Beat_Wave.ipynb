```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# time array
t = np.linspace(0,0.1,1000,endpoint=False)
```

```python
# low Frequency
Low_HZ = 30
# High Frequency
High_HZ = 220
```

```python
# define Sinusoid
Sinusoid = np.cos(2*np.pi*Low_HZ*t) * np.cos(2*np.pi*High_HZ*t)
```

```python
env = np.cos(2*np.pi*Low_HZ*t)
env2 = -np.cos(2*np.pi*Low_HZ*t)
```

```python
# plotting 
plt.plot(t,Sinusoid,'-')
plt.plot(t,env,'--',color='b')
plt.plot(t,env2,'--',color='g')

plt.xlabel('t(second)')
plt.ylabel('Amplitude')
plt.axis([ 0, 0.1, -1, 1 ])
plt.show()
```
