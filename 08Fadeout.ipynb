```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# time array
t = np.linspace(0,1,1000,endpoint=False)
```

```python
# define Sinusoid
Sinusoid = np.cos(2*np.pi*5*t)
```

```python
# define fadeout
Fadeout = np.linspace(1,0,1000,endpoint = False)
```

```python
Sinusoid = Sinusoid * Fadeout
```

```python
# plotting
plt.plot(t,Sinusoid)

plt.xlabel('t(second)')
plt.ylabel('Amplitude')
plt.axis([ 0, 1, -2, 2 ])
plt.show()
```
