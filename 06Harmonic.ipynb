```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# time array
t = np.linspace(0,1,1000,endpoint=False)
```

```python
# define Fundamental Frequency
Base_F = 2
# define first Sinusoid
x1 = np.cos(2*np.pi*Base_F*t)
# define second Sinusoid
x2 = np.cos(2*np.pi*4*Base_F*t)
# Combine two Sinusoid
x3 = x1 + x2 
```

```python
# plotting x1
plt.figure(1)
plt.plot(t,x1)
plt.xlabel('t(second)')
plt.ylabel('Amplitude')
plt.axis([ 0, 1, -2, 2 ])
plt.show()
```

```python
# plotting x2
plt.figure(2)
plt.plot(t,x2)
plt.xlabel('t(second)')
plt.ylabel('Amplitude')
plt.axis([ 0, 1, -2, 2 ])
plt.show()
```

```python
# plotting x3
plt.figure(3)
plt.plot(t,x3)
plt.xlabel('t(second)')
plt.ylabel('Amplitude')
plt.axis([ 0, 1, -2, 2 ])
plt.show()
```
